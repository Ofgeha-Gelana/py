```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('canada_per_capita_income.csv')
df.head(5)
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('capita_income')
plt.scatter(df.year,  df['per capita income (US$)'], color='red',marker='+')
```

```python
new_df = df.drop('per capita income (US$)', axis='columns')
new_df.head(5)
```

```python
capita_per_income = df['per capita income (US$)']
capita_per_income.head(5)
```

```python
# Create linear regression object
reg = linear_model.LinearRegression()
reg.fit(new_df,capita_per_income)
```


## (1) Predict capta_per_income of a home with year = 2020 

```python
reg.predict([[2020]])
```

```python
reg.coef_
```

```python
reg.intercept_
```
